```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Define the sets
dont_watch_games = set(['Alice', 'Bob', 'Charlie', 'David'])
unhappy_with_joe = set(['Charlie', 'David', 'Eve', 'Frank'])

# Create the Venn diagram without displaying set sizes
venn_diagram = venn2([dont_watch_games, unhappy_with_joe], set_labels=('Don\'t watch games', 'Unhappy with Joe'))

# Add title
plt.title("Venn Diagram: Joe Sucks")

# Hide the numbers inside the circles
venn_diagram.get_label_by_id('10').set_text('')
venn_diagram.get_label_by_id('11').set_text('')
venn_diagram.get_label_by_id('01').set_text('')

# Add set labels inside the diagram
plt.text(-0.7, -0.3, "Don't watch games", fontsize=10, ha='center', va='center')
plt.text(0.7, -0.3, 'Unhappy with his press conference', fontsize=10, ha='center', va='center')

# Add text in the middle
plt.text(0, 0, 'JOE SUCKS', color='red', fontsize=12, ha='center', va='center')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define data
labels = ['Mazzulla ball rulez babybaby', 'I only like Timeouts']
sizes = [97.74, 2.26]  # Sizes in percentage

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.2f%%', colors=['skyblue', 'lightcoral'])

# Add title
plt.title('Celtics fans')

# Show plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
