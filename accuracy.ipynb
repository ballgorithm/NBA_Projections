```python
import pandas as pd
from prettytable import PrettyTable
from tabulate import tabulate
```

```python
df = pd.read_excel('future.xlsx', sheet_name='Sheet2')
```

```python
df
```

```python
print("\nResult Table:")
print(tabulate(df[['Date', 'Accuracy']], headers='keys', tablefmt='pretty'))
```

```python
df = df.fillna('')
```

```python
df['Accuracy'].fillna('', inplace=True)

# Convert "Accuracy" column to numeric values
df['Accuracy'] = pd.to_numeric(df['Accuracy'], errors='coerce')

# Format the "Accuracy" column to display 2 decimal places or leave as blank
df['Accuracy'] = df['Accuracy'].apply(lambda x: '\033[92m{:.2f}\033[0m'.format(x) if pd.notna(x) else '')

# Display the DataFrame using tabulate with "Accuracy" values formatted to 2 decimal places and green text
print("\nResult Table:")
print(tabulate(df[['Date', 'Accuracy']], headers='keys', tablefmt='pretty', missingval='', showindex=False))
```
